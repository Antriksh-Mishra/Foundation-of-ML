PART F — Memory Safety
Task F1 — Copy VS View

1. Create a reference backup:
2. Create a safe backup using:
3. safe_backup = electricity.copy()
4. Modify one value in the original electricity array.
5. Observe which backup changes and which remains unchanged.

Explain:
1. Why is ref_backup risky?
2. When is .copy() required?

```python
import numpy as np

electricity = np.random.randint(120,651,(4,7))

# Step 1: Reference backup (view)
ref_backup = electricity

# Step 2: Safe backup (independent copy)
safe_backup = electricity.copy()

# Step 3: Modify original array
electricity[0,0] = 9999

print("Original Electricity:\n", electricity)
print("\nReference Backup:\n", ref_backup)
print("\nSafe Backup:\n", safe_backup)
```

```python
# Explanation 
# ref_backup = electricity
# → Both variables point to the same memory location.
# → Modifying one modifies the other.

# safe_backup = electricity.copy()
# → Creates a completely new independent array.
# → Safe from accidental modification.

#  When is .copy() Required?

# When preserving original data before modification

# Before applying filters or transformations

# When passing arrays to functions that may modify them
```
