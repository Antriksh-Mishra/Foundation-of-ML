Axis in NumPy

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print(np.sum(arr)) # sum of all elements
print(np.sum(arr, axis=0)) # sum of each column 
print(np.sum(arr, axis=1)) # sum of each row
```

"Where" in NumPy

```python
arr=np.array([1,-2,3,-4,5])
result1 = np.where(arr<0,0,arr) # replace negative numbers with 0
print(result1)
result2 = np.where(arr % 2!=0,-1,arr) # replace even numbers with -1
print(result2)
```

```python
data=np.array([10,20,30,40,50,60,70,80,90])
mean_val=np.mean(data) # mean of all elements
result=np.where(data>mean_val,mean_val,data)# replace elements greater than mean with mean value
print(result)
```

Applying "Where" condition in 2D array

```python
mat=np.array([[10,20,30],[40,50,60],[70,80,90]])
result = np.where(mat>=50,1,0) # binary mask for elements greater than or equal to 50
print(result)               
```
