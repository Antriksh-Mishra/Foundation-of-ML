```python
data = {
    "Name" : ["Aman","Riya", "Rahul", "Sneha","Arjun","Priya","karan","Neha"],
    "Age": [22,23,25,24,26,27,28,22],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female",],
    "Marks": [99,78,95,63,97,81,46,69]
}
print(data)
```

Line plot for Marks trend

```python
plt.plot(data["Marks"])
plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Bar plit for comparision

```python
plt.bar(data["Name"],data["Marks"])
plt.title("Marks of Student")
plt.xlabel("Students Name")
plt.ylabel("Marks")
plt.show()
```

Scatter PLot 

```python
plt.scatter(data["Age"], data["Marks"])
plt.show()
```

Gender Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Name": ["Antriksh", "Riya", "Rahul", "Sneha", "Amit"],
    "Gender": ["Male", "Female", "Male", "Female", "Male"]
})

gender_counts = data["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```
